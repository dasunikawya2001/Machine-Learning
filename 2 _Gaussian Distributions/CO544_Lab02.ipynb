```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Dot product of x and y
x = np.array([1, 2])
y = np.array([-2, 1])
a = np.dot(x, y)
# Answer = 0 , those two are orthogonal
print(a)

# Norm
# Both functions compute the Euclidean Norm
b = np.linalg.norm(x)
c = np.sqrt(x[0]**2 + x[1]**2)
print(b, c)

# This gives the angle between x and y
theta = np.arccos(np.dot(x,y) / (np.linalg.norm(x) * np.linalg.norm(x)))
print("Angle between x and y:" , math.degrees(theta))

# Symmetric matrix
# B = Transpose of B
# Hence, B - B.T =0
B = np.array([[3,2,1], [2,6,5], [1,5,9]], dtype=float)
print(B)
print(B- B.T)

# Generate z with random 3 values
z = np.random.rand(3)
# Matrix vector multiplication
v = B @ z
# Shape of v is 3.
print(v.shape)

# Quadratic form
# z.T - get the transpose of the vector z
# B @ Z This is a vector multiplication
# z.T @ (B @ z) is a dot product between two vectors
# It gives a single number.
print(z.T @ B @ z)

# Trace - sum of diagonal elements of a matrix
print(np.trace(B))

# Computes the determinent of the matrix B
print(np.linalg.det(B))

# Eigen values and Eigen Vectors
D, U = np.linalg.eig(B)
print(D) #Eigenvalues
print(U) #Eigenvectors

# Computes the dot product of the first two eigenvectors of B
# Eigenvectors of a symmetric matrix are orthogonal.
# Since B is symmetric, output is expected to be zero.
# The output is approximately 0.
# This confirms that the eigenvectors are orthogonal.
print(np.dot(U[:,0], U[:,1]))


# Here multiply eigen vector by its transform
# Out will be Identity matrix, because this is a eigen vector of a symetric matrix
print(U @ U.T)
```

```python
MaxTrial = 2000
sampleSizeRange = np.linspace(100, 200, 40)
plotVar = np.zeros(len(sampleSizeRange))

for sSize in range(len(sampleSizeRange)):
    numSamples = int(sampleSizeRange[sSize])
    vStrial = np.zeros(MaxTrial)

    for trial in range(MaxTrial):
        xx = np.random.randn(numSamples, 1)  # Normal samples
        vStrial[trial] = np.var(xx)         # Variance of each trial

    plotVar[sSize] = np.var(vStrial)        # Variance of all 2000 variances

# Plotting the result
fig, ax = plt.subplots(figsize=(4, 4))
ax.plot(plotVar)
```

```python
def gauss2D(x, m, C):
    Ci = np.linalg.inv(C)         # Inverse of covariance matrix
    dC = np.linalg.det(C)         # Determinant (used in normalization)
    num = np.exp(-0.5 * np.dot((x - m).T, np.dot(Ci, (x - m))))  # Numerator of PDF
    den = 2 * np.pi * np.sqrt(dC) # Denominator of PDF
    return num / den              # Gaussian probability density

def twoDGaussianPlot(nx, ny, m, C):
    x = np.linspace(-5, 5, nx)   # Create grid values for X
    y = np.linspace(-5, 5, ny)   # Create grid values for Y
    X, Y = np.meshgrid(x, y, indexing='ij')  # Create 2D coordinate grid
    Z = np.zeros([nx, ny])
    for i in range(nx):
        for j in range(ny):
            xvec = np.array([X[i, j], Y[i, j]])    # Each 2D point
            Z[i, j] = gauss2D(xvec, m, C)          # Calculate PDF value
    return X, Y, Z

nx, ny = 50, 40
m1 = np.array([0, 2])                        # Mean of distribution
C1 = np.array([[2, 1], [1, 2]], np.float32)  # Covariance matrix
Xp, Yp, Zp = twoDGaussianPlot(nx, ny, m1, C1)
plt.contour(Xp, Yp, Zp, 5)                   # Plot 5 contour levels
```

```python
# Parameters for each distribution
params = [
    (np.array([2.4, 3.2]), np.array([[2, -1], [-1, 2]]), 'Dist 1'),
    (np.array([1.2, 0.2]), np.array([[2, 0], [0, 4]]), 'Dist 2'),
    (np.array([2.4, 3.2]), np.array([[2, 0], [0, 2]]), 'Dist 3')
]

# Plot
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, (mean, cov, title) in enumerate(params):
    X, Y, Z = twoDGaussianPlot(100, 100, mean, cov)
    axs[i].contour(X, Y, Z, levels=5)
    axs[i].set_title(title)
    axs[i].set_aspect('equal')
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the covariance matrix and generate samples
C = np.array([[2.0, 1.0], [1.0, 2.0]])        # Covariance matrix
A = np.linalg.cholesky(C)                    # Cholesky decomposition
X = np.random.randn(1000, 2)                 # Standard normal samples
Y = X @ A                                    # Transform to match desired covariance

# Step 2: Sweep theta from 0 to 2π and compute projected variances
nPoints = 100
thRange = np.linspace(0, 2 * np.pi, nPoints)
pVars = np.zeros(nPoints)

for i, theta in enumerate(thRange):
    u = np.array([np.sin(theta), np.cos(theta)])
    yp = Y @ u
    pVars[i] = np.var(yp)

# Step 3: Plot projected variances as function of theta
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(thRange * 180 / np.pi, pVars)
ax.set_title("Variance of Projected Data vs Angle θ (in degrees)")
ax.set_xlabel("θ (degrees)")
ax.set_ylabel("Variance")
plt.grid(True)
plt.tight_layout()
plt.show()
```
